```python
class TrafficDevice:
    def __init__(self, device_id):
        self.device_id = device_id

    def activate(self):
        print(f"[{self.device_id}] Generic device activated.")

class TrafficLight(TrafficDevice):
    def activate(self):
        print(f"[{self.device_id}] Traffic light cycling: RED -> YELLOW -> GREEN.")

class SpeedCamera(TrafficDevice):
    def activate(self):
        print(f"[{self.device_id}] Speed camera scanning passing vehicles for speed violations.")


class PedestrianSignal(TrafficDevice):
    def activate(self):
        print(f"[{self.device_id}] Pedestrian signal showing WALK sign.")

class EmergencySiren(TrafficDevice):
    def activate(self):
        print(f"[{self.device_id}] Emergency siren is blaring")

def main():
    # One object of each class
    devices = [
        TrafficLight("TL1"),
        SpeedCamera("SC1"),
        PedestrianSignal("PS1"),
        EmergencySiren("ES1"),
    ]
    for device in devices:
        device.activate()

if __name__ == "__main__":
    main()
```
